```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate Spatial Coordinates
np.random.seed(0)
num_locations = 100
min_lat, max_lat = -1.0, 1.0
min_lon, max_lon = -1.0, 1.0
coordinates = np.random.uniform(low=(min_lat, min_lon), high=(max_lat, max_lon), size=(num_locations, 2))

# Step 2: Generate Environmental Covariates
def generate_covariates(coordinates):
    # Example: Simulate temperature and humidity as covariates
    temperature = norm(loc=25, scale=3).pdf(coordinates[:, 0])
    humidity = norm(loc=0.6, scale=0.1).pdf(coordinates[:, 1])
    return temperature, humidity

temperature, humidity = generate_covariates(coordinates)

# Step 3: Simulate True EIR using Garki Model
def simulate_eir(temperature, humidity):
    # Example: Simple linear model for EIR based on temperature and humidity
    beta_temp = 0.5
    beta_humidity = 0.3
    true_eir = beta_temp * temperature + beta_humidity * humidity + np.random.normal(loc=0, scale=0.1, size=len(temperature))
    return true_eir

true_eir = simulate_eir(temperature, humidity)

# Step 4: Convert to Age-Specific Infection Prevalence
def eir_to_infection_prevalence(true_eir):
    # Example: Simple conversion to age-specific infection prevalence
    age_groups = ['0-5', '6-10', '11-15', '16-20', '21+']
    prevalence = np.random.uniform(low=0.1, high=0.9, size=(len(age_groups), len(true_eir)))
    return age_groups, prevalence

age_groups, infection_prevalence = eir_to_infection_prevalence(true_eir)

# Plotting the results
plt.figure(figsize=(10, 6))
for i in range(len(age_groups)):
    plt.plot(true_eir, infection_prevalence[i], label=f'Age Group: {age_groups[i]}')
plt.xlabel('True EIR')
plt.ylabel('Infection Prevalence')
plt.title('Age-Specific Infection Prevalence vs. True EIR')
plt.legend()
plt.grid(True)
plt.show()
```
